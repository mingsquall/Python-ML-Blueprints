# 使用Gutenberg语料库

```python
print(gutenberg.fileids()) # 计算nltk目前收录的gutenberg项目中收录的书
```

```python
allwords = gutenberg.words('shakespeare-hamlet.txt')
```

```python
len(allwords)
```

```python
len(set(allwords))
```

```python
allwords.count('Hamlet')
```

```python
A = set(allwords)
```

```python
longwords = [w for w in A if len(w) > 12]
```

```python
print(sorted(longwords))
```

```python
# freqG20
```

```python
fd2 = FreqDist([sx.lower() for sx in allwords if sx.isalpha()])
# FreqDist()用来创建所给数据的频率分布
```

```python
print(fd2.B())
# 计算不同单词的个数
```

```python
print(fd2.N())
# 计算所有单词的个数
```

```python
fd2.tabulate(20)
# 前20组数据以表格的形式显示出来
```

```python
fd2.plot(20)
```

```python
fd2.plot(20, cumulative=True)
```

# 使用就职演说语料库

```python
fd3 = FreqDist([s for s in inaugural.words()])
print(fd3.freq('freedom'))
```

```python
cfd = ConditionalFreqDist((fileid, len(w))
    for fileid in inaugural.fileids()
    for w in inaugural.words(fileid)
    if fileid > '1980' and fileid < '2010')
print(cfd.items())
cfd.plot()
```
