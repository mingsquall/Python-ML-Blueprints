```python
import tushare as ts
import matplotlib.pyplot as plt
import seaborn as sns
import seaborn.regression as snsl
from datetime import datetime
```

```python
sns.set_style("whitegrid")
```

```python
end = datetime.today()
start = datetime(end.year-1, end.month, end.day)
# 开始时间/结束时间，选取最近1年的数据
end = str(end)[:10]
start = str(start)[:10]
```

```python
stock = ts.get_hist_data('600028',start=start,end=end)
```

## 收盘价走势图

```python
stock['close'].plot(legend=True, figsize=(10,4)) # 添加图例，设置size大小是宽10英寸，高4英寸
```

```python
plt.show()
```

## 收盘价、5日均线、10日均线以及20日均线

```python
stock[['close','ma5','ma10','ma20']].plot(legend=True,figsize=(10,4))
```

```python
plt.show()
```

```python
## 股票每日涨幅读
```

```python
stock['Daily Return'] = stock['close'].pct_change()
```

```python
stock['Daily Return'].plot(legend=True,figsize=(15,5))
```

```python
plt.show()
```
